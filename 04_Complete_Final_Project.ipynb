 ##  Important features

The most important columns in the secondary dataset are AI usage hours, commits, and coding hours. These columns show what developers are doing and how well they are doing it. AI usage hours show how often developers use AI tools, coding hours show how long they spend on development work, and commits show how much work they have done. These features, when combined, give a clear picture of how productive and efficient someone is.

The most important parts of the primary dataset are tweets and labels because they show how people feel about ChatGPT. The tweets have the actual text data that shows what users are thinking, and the labels show how they feel (positive, negative, or neutral). These features are necessary for sentiment analysis and for figuring out how people feel about AI tools.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df1['tweets'])

y = df1['labels']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

### Justification and discussion on sentiment analysis model 
For sentiment analysis on the main dataset in this project, we used the Naive Bayes (MultinomialNB) model. We chose this model because it is simple, quick, and very good at classifying text, like figuring out how someone feels. It works well with a lot of text data and is often used to sort text into groups like positive, negative, and neutral.

Count Vectorizer changed the text data (tweets) into numbers before the model was used. This step changes words into numbers based on how often they are used, which helps the model understand and process the text. After that, the dataset was split into training and testing sets so that the model could be tested.

The Naive Bayes model is simple but still works well for many real-world uses because it assumes that words are not related to each other. It worked well for us and gave a good accuracy score, which shows that it can correctly figure out how users feel.

Overall, this model was a good fit for our project because it is easy to use, doesn't need as much processing power, and works well for sentiment classification tasks.

```python
# split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

In this project, we chose three machine learning models, namely, Linear Regression, Decision Tree, and Random Forest, depending on the character of the dataset and the type of problem.

The reason why Linear Regression was used is that it is easy to use and it is applicable in numerical data. It assists in determining the relationship between the variables used as inputs (hours of AI use and hours of code) and the target variable (commits). It is appropriate in case the data presents a linear relationship.

Decision Tree was chosen as it is able to deal with both simple and complicated relationships in the data. It does not need any linear relationship and can detect the patterns by dividing the data into various branches. It can also be interpreted and comprehended easily.

Random Forest was selected because it is a developed model that integrates several decision trees to enhance the accuracy and minimize errors. It is suitable with sophisticated data and prevents overfitting. It helps to make more reliable predictions as opposed to a single decision tree.

These models were suitable to our dataset since it has numerical features and the task to predict developer activity and therefore regression based models are suitable to this task

# Model 1: Linear Regression

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)

pred1 = lr.predict(X_test)

print("LR RMSE:", mean_squared_error(y_test, pred1))
```

# Model 2: Decision tree

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)

pred2 = dt.predict(X_test)

print("DT RMSE:", mean_squared_error(y_test, pred2))
```

# Model 3: Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

pred3 = rf.predict(X_test)

print("RF RMSE:", mean_squared_error(y_test, pred3))
```

# Evaluation of Models

### Evaluation Technique Used

We used RMSE (Root Mean Squared Error) to see how well the machine learning models worked in this project. The square root of the average of the squared errors is used to find the RMSE, which shows how far off the predicted values are from the real values. This means that it gives bigger mistakes more weight, which is helpful for figuring out how well the model works.

RMSE gives you one number that shows how wrong the model's predictions were overall. The model is doing a better job if the RMSE value is lower because it means the predicted values are closer to the real values. A higher RMSE value, on the other hand, means bigger mistakes and worse performance.

RMSE is simple to understand and is often used for regression problems with a numerical target variable. So, it is a good and useful way to measure how well our machine learning models work in this project.

### Application of Evaluation to Each Model

The RMSE value was calculated for each model to compare their performance:

Linear Regression RMSE: 5.41

Decision Tree RMSE: 10.48

Random Forest RMSE: 5.65

These values show that Linear Regression has the lowest error, followed by Random Forest, while Decision Tree has the highest error. This comparison helps us understand which model performs better on the dataset.

### Appropriate Metric for Model Type
Because the target variable (commits) is a number that can change over time, RMSE is a good way to measure how well this project is going. RMSE is a good choice for a regression problem because it measures the difference between the actual and predicted values in the same unit as the target variable. This makes it easy to understand and make sense of the results.

RMSE also gives more weight to bigger mistakes, which helps find models that make big prediction mistakes. This is helpful because in real life, big mistakes can have a bigger effect than small ones.

RMSE is a clear and reliable way to measure how well a model works for regression tasks compared to other metrics. So, it's the right and useful way to measure how well these models work in our project.
